#### 1. Sum of Beauty of All Strings -> Leetcode 1781

O(n^2) is the best solution. When we are talking about substrings, two pointer approach is generally not used. 
You'll have to find out all substrings by nested loop. 

```python
s = "aabcb"
N = len(s)
# tmp_str = str()
temp = dict()
ans = 0

for i in range(0,N):
    for j in range(i,N):
        if s[j] in temp:
            temp[s[j]] += 1
        else:
            temp[s[j]] = 1


        ans += max(temp.values()) - min(temp.values())

    temp.clear()

print(ans)
```

#### 3. Sherlock and Anagrams -> Hackerrank

```python
s = 'cdcd'

# anagrams, are two strings, when rearranged can be converted into each other, 'mom' and 'mmo', that's it. 

# 2 strings are anagrams if there frequency map is same. 3P2 -> 3!/2!
# abba, 1, 1, 2, nPr / 

tmp_str = str()
frequency_map = dict()

# Find all substrings. 
for i in range(0,len(s)):
    for j in range(i,len(s)):
        tmp_str = tmp_str + s[j]
        tmp_str = ''.join(sorted(tmp_str))

        if tmp_str in frequency_map:
            frequency_map[tmp_str] += 1
        else:
            frequency_map[tmp_str] = 1

    tmp_str = ''

print(frequency_map)

count = 0
for _, values in frequency_map.items():
    # if values
    count += (values * (values - 1)) // 2

print(count)


# we take out all substrings, sort, and store same substrings. Now whatever is number of same substrings, total
# no. of pairs for these 3 similar substrings, nC2,  -> n! / (n-2)!2! ----> n * n - 1 / 2
# no. of all possible pairs of substrings, will be count of anagrams, it will be anagram since substrings are same, 
# have same fequency map. 
```

#### 4. Reverse Words in a String 151. -> Leetcode

```python
s = "  hello world  "
word = []
tmp_str = str()

for i in range(0,len(s)):
    tmp_str += s[i] if s[i] != ' ' else ''
    if s[i] == ' ' or i == len(s) - 1:
        word.append(tmp_str)
        tmp_str = ''
    
    
word = [x for x in word if x != '']

print(word)
print(' '.join(word[::-1]))
```

```python
# Once more

# Break sentence into line tokens. 
# words = ["This", "is", "an", "example", "of", "text", "justification."]
# words = ["What","must","be","acknowledgment","shall","be"]
words = ["Science","is","what","we","understand","well","enough","to","explain","to","a","computer.","Art","is","everything","else","we","do"]

maxWidth = 20

def line_tokens(words: list[str], maxWidth: int) -> list[list[str]]:
    N = len(words)
    count = int()
    tmp_list = list()
    ans = list()
    i = 0

    while ( i < N ):
        tmp_list.append(words[i])
        count += len(words[i])

        if count > maxWidth:
            tmp_list.pop()
            count = 0
            ans.append(tmp_list.copy())
            tmp_list.clear()
        else:
            i += 1
            count += 1

        if i == N:
            ans.append(tmp_list.copy())
            tmp_list.clear()

        # print(ans, tmp_list)


    return ans

tokens = line_tokens(words, maxWidth)
# print(tokens)




# now we make justified sentences, break into two types, when sentence will be left justified. 
def left_justified(token: list[str], maxWidth: int) -> list[str]:
    count = 0
    tmp_str = str()
    for t in token:
        tmp_str += t
        tmp_str += " "
        count += len(t) + 1

# striping end for better padding. Se can avoid edge cases where last word is followed by a space as from above.
# works in single word sentence, not in last complete sentence. 
# AGAIN, BREAK DOWN PROBLEM AND CONQUER ALL !
    count -= 1
    tmp_str = tmp_str.rstrip()

    padding = maxWidth - count
    tmp_str += " "*padding
    return tmp_str


# when sentences are right justified. 
def equal_justified(token: list[str], maxWidth: int) -> list[str]:
    no_of_charaters = lambda token: sum(len(word) for word in token)
    padding = maxWidth - no_of_charaters(token)
    slots = len(token) - 1
    equal_padding = padding // slots
    remaining = padding % slots
    tmp_str = str()

    for i in range(0, len(token)):
        tmp_str += token[i]
        tmp_str += " "*equal_padding
        tmp_str += " " if remaining else ""
        remaining -= 1 if remaining else 0 

    tmp_str = tmp_str.rstrip()
    return tmp_str






ans_list = list()

for i in range(0, len(tokens)):

    if i == len(tokens) - 1 or len(tokens[i]) == 1:
        ans_list.append(left_justified(tokens[i], maxWidth))
    else:
        ans_list.append(equal_justified(tokens[i], maxWidth))



for word in ans_list:
    print(word, len(word))
# Left justified if one word in a line or last line.
# Equal justified if other cases. 
```
