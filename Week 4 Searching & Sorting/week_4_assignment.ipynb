### 1. K-diff in an array leetcode - 532

```python
# First sorting, O(nlong), then going to each element, getting sum with target and searching if pair exisit with bs
# Two pointer approach with pointers starting at 1st and 2nd element


nums = [3,1,4,1,5]
k = 2
nums.sort() # O(nlogn)
        
N = len(nums)
if N < 2:
    print(k)
i = 0
j = 1
ans = set()

#O(n)

while ( j != N ):
    

    if nums[j] - nums[i] == k:
        ans.add((nums[j],nums[i]))
        i += 1
        j += 1
    elif nums[j] - nums[i] > k:
        i += 1
    elif nums[j] - nums[i] < k:
        j += 1
    if i == j:
        j += 1


print(ans)
print(int(len(ans)))
```

### 4. Book Allocation Problem

```python
# Kind of like cows. 

arr = [15, 10, 19, 10, 5, 18, 7]
k = 5

def predicate(arr: list[int], k: int, pages: int) -> bool:
    books_counter = 0
    i = 0
    pages_alloted = 0

    while(k > 0 and i < len(arr)):
        
        #allocating
        pages_alloted += arr[i]
        if pages_alloted > pages:
            k -= 1
            pages_alloted = 0
        else:
            books_counter += 1
            i += 1
        print(books_counter, pages_alloted, k)
    k -= 1
    if k != 0 and books_counter != len(arr):
        return False
    else:
        return True


start = 0
end = sum(arr)
# print(j)
ans = -1
# if more students than books, then return -1
if k > len(arr):
    print(-1)


while (start <= end):
    mid = start + (end - start)//2
    print('mid: ',mid)
    if predicate(arr, k, mid):
        ans = mid
        end = mid - 1
    else:
        start = mid + 1
    print('ans:', ans)


print(ans)
```

### 5. The Painter's Partition Problem - II

```python
# Kind of like cows. 

arr = [21, 17, 3, 12, 21, 8, 11, 4, 8, 22, 24]
k = 6
ans = -1

def predicate(arr: list[int], k: int, mid: int) -> bool:
    i = 0
    painters_progress = 0

    while ( i < len(arr) and k > 0):
        
        painters_progress += arr[i]

        if mid >= painters_progress:
            i += 1
        else:
            k -= 1
            painters_progress = 0

        print(painters_progress, i, k)

    if i != len(arr):
        print('false', i, len(arr))
        return False
    else:
        print('true', i, len(arr))
        return True


start = 0
end = sum(arr)

while (start <= end):
    mid = start + (end - start)//2

    if predicate(arr, k, mid):
        ans = mid
        end = mid - 1
    else:
        start = mid + 1
    print(mid)

print(ans)
```

### 6. Aggresive Cows

Needs to be sorted, then predicate to see possilbility. I return false only when cows are left, if array is left but 
cows are distributed, it's a possible ans, now to maximize store and compute.

```python
stalls = [1, 2, 4, 8, 9]
k = 3

def predicate(arr: list[int], cows: int, minimum_distance: int) -> bool:
    # No. of available stalls
    N = len(arr)
    arr.sort()
    # If less stalls than cows, not possible condition
    if N < cows:
        return False
    
    # Getting and counter to traverse through stalls
    i = 0

    # Placing first cow at zero position for initialization. since from there we will evaluate difference 
    cows -= 1
    i += 1
    j = 0

    while ( i < N and cows > 0 ):

        # checking if can keep cow at current stall
        if minimum_distance <= abs(arr[i] - arr[j]):
            cows -= 1
            j = i
            i += 1 
        else:
            i += 1

    
    if cows != 0:
        return False
    
    return True





start = 0
end = max(stalls)
ans = -1

while (start <= end):
    mid = start + (end - start)//2

    if predicate(stalls, k, mid):
        # ans = min(ans,mid)
        ans = mid
        start = mid + 1
    else:
        end = mid - 1


print(ans)
```
