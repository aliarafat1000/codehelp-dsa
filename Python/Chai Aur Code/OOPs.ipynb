# Magic Methods for OOPs

Magic methods main magic is that they can call themselves, we don't need to call them. Like __init__ called itself, 
we didn't needed to call it.

__str__ method, we can use print(object) this method is calls

__slots__ can put which attributes are allowed in classes, no attributes other than this can be made. _Class__attr mangles name to put private attributes like __attr. Attempting to add new attributes, object.new_attr = var, not allowed any more. Otherwise it's allowed. To inherit __slots__ also declare it in child class, otherwise it won't work in child class.

__add__ (self, other), ->  (arg1, arg2) but there is no self or other. It's just naming convention

__sub__ -> the same as above

__mul__ -> the same as above

__truediv__ -> same as above

__eq__, __gt__ etc are like above as well

```python
class Car:

    total_car = 0
    __slots__ = ['_Car__brand', '_Car__model']
    
    def __init__(self, brand, model):
        self.__brand = brand
        self.__model = model
        Car.total_car += 1 # Can use self.total_car as well 

    def full_name(self):
        return f'{self.__brand} {self.__model}'
    

    #### magic methods
    def __str__(self):
        return f'Straight from magic method {self.__brand} {self.__model}'
    
    def __add__(self, other):
        return Car(self.__brand + other.__brand, self.__model + other.__model)
        # return "I am trying to add, but I don't know"

    def __mul__(self,other):
        return Car(self.__brand*2 + other.__brand*2, self.__model + other.__model)





class ElectricCar(Car):
    __slots__ = ['battery_size']
    def __init__(self, brand, model, batter_size):
        super().__init__(brand, model)
        self.battery_size = batter_size

    def fuel_type(self):
        return "Electric Charge"
    

my_tesla = ElectricCar('Tesla', 'Model Y', '85 KWH')
print(my_tesla)
my_car = Car('Tata', 'Harrier')
yo = my_tesla + my_car
yo.full_name()

yes = (my_tesla*my_car)
yes.full_name()
```
