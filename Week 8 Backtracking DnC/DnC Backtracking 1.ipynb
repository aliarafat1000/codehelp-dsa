# Merge two sorted arrays

````python
num1 = [1,5,8,13]
num2 = [2,4,12,20,25]
ans = []

i, j = 0, 0

while (i < len(num1) and j < len(num2)):
    if num1[i] > num2[j]:
        ans.append(num2[j])
        j += 1
    else:
        ans.append(num1[i])
        i += 1


if i < len(num1):
    ans.extend(num1[i:])
else:
    ans.extend(num2[j:])

print(ans)
````

```markdown
## â±ï¸ Merge Sort Time Complexity

### 1. How Merge Sort Works
1. **Divide** the array into 2 halves â†’ until each part has 1 element.  
2. **Merge** the two sorted halves back into one sorted array.  

---

### 2. Depth of Recursion (How many times we divide)
- Each time the array size halves.  
- Example for 8 elements:  
```

8 â†’ 4 + 4 â†’ 2+2+2+2 â†’ 1+1+1+1+1+1+1+1

```
- Number of divisions = **logâ‚‚(n)**.  

So recursion depth = **O(log n)**.  

---

### 3. Cost of Merging
- At each recursion level, all `n` elements are merged.  
- Merging cost = **O(n)**.  

---

### 4. Total Work
- Work per level = **O(n)**  
- Number of levels = **O(log n)**  

\[
\text{Total} = O(n) \times O(\log n) = O(n \log n)
\]

---

### âœ… Final Result
- **Best Case**: O(n log n)  
- **Average Case**: O(n log n)  
- **Worst Case**: O(n log n)  
- **Space Complexity**: O(n) (extra arrays used during merging)  

---

âš¡ **Shortcut to Remember**:  
ðŸ‘‰ *In Merge Sort, every element is touched at every level of recursion â†’ `n log n`.*
```
